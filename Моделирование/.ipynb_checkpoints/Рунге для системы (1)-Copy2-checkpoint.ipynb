# Рунге для 2-ой системы

```python
import numpy as np
import math
size = 7           # размерность системы
```

```python
# начальные данные
S_b = 0.0001   # квадратный сантиметр
p_atm = 1e+5
p_a = p_atm
p_b = p_atm
S_a = S_b
m = 0.1   # 100 грамм
q = 1e-3
c = 1e-9

B = 1
g = 9.81
mu = 0.09
gam = 9.81

ck = c
```

```python
sign = lambda x: (1, -1)[x < 0]
sign(-5), sign(5)
```

```python
def fun(t, y, qA, qB):
    k = [0] * (size - 3)
    k[0] = y[1]
    k[1] = (y[3] * S_b - y[2] * S_a) / m
    k[2] = (qA + y[1] * S_a) / c
    k[3] = (qB - y[1] * S_b) / c
    return k
```

```python
def fun2(t, y, qA):    # t, y = const
    k = (q - qA) / c
    return k
```

```python
def fun3(t, qA, pA, pa, S): # pA - вход, pa - выход
    qA_next = B * (mu * S * sign(pA - pa) * math.sqrt(2 * g / abs(pA - pa) / gam) - qA)
    return qA_next
```

```python
def Runge(t, y, h):  # t - значение узла, y - значения в точке t
    # коэффициенты Рунге
    k = [[0] for _ in range(4)]
    y_temp = [0] * (size - 3)
    # значения 7 функций на следующем шаге
    y_result = [0] * size
    # копия в текущем узле (для перемен)
    y_local = y.copy()
    
    k[0] = fun2(t, y_local[5], y_local[4])
    k[1] = fun2(t + h / 2, y_local[5] + h * k[0] / 2, y_local[4])
    k[2] = fun2(t + h / 2, y_local[5] + h * k[1] / 2, y_local[4])
    k[3] = fun2(t + h, y_local[5] + h * k[2], y_local[4])
    kn = (k[0] + 2 * k[1] + 2 * k[2] + k[3]) / 6
    # новое значение pA
    y_result[5] = y_local[5] + h * kn
    y_local[5] += h * kn

    k[0] = fun3(t, y_local[4], y_local[5], y_local[2], S_a)
    k[1] = fun3(t + h / 2, y_local[4] + h * k[0] / 2, y_local[5], y_local[2], S_a)
    k[2] = fun3(t + h / 2, y_local[4] + h * k[1] / 2, y_local[5], y_local[2], S_a)
    k[3] = fun3(t + h, y_local[4] + h * k[2], y_local[5], y_local[2], S_a)
    kn = (k[0] + 2 * k[1] + 2 * k[2] + k[3]) / 6
    # новое значение qA
    y_result[4] = y_local[4] + h * kn
    y_local[4] += h * kn

    k[0] = fun(t, y_local, y_local[4], y_local[6])

    for i in range(size - 3):
        y_temp[i] = y_local[i] + h / 2 * k[0][i]
    k[1] = fun(t + h / 2, y_temp, y_local[4], y_local[6])

    for i in range(size - 3):
        y_temp[i] = y_local[i] + h / 2 * k[1][i]
    k[2] = fun(t + h / 2, y_temp, y_local[4], y_local[6])

    for i in range(size - 3):
        y_temp[i] = y_local[i] + h * k[2][i]
    k[3] = fun(t + h, y_temp, y_local[4], y_local[6])

    for i in range(size - 3):
        y_result[i] = y_local[i] + h * (k[0][i] + 2 * k[1][i] + 2 * k[2][i] + k[3][i]) / 6
        y_local[i] += h * (k[0][i] + 2 * k[1][i] + 2 * k[2][i] + k[3][i]) / 6

    k[0] = fun3(t, y_local[6], y_local[3], p_atm, S_b)
    k[1] = fun3(t + h / 2, y_local[6] + h * k[0] / 2, y_local[3], p_atm, S_b)
    k[2] = fun3(t + h / 2, y_local[6] + h * k[1] / 2, y_local[3], p_atm, S_b)
    k[3] = fun3(t + h, y_local[6] + h * k[2], y_local[3], p_atm, S_b)
    kn = (k[0] + 2 * k[1] + 2 * k[2] + k[3]) / 6
    # новое значение qB
    y_result[6] = y_local[6] + h * kn
    return y_result
```

```python
def hydraulic(t_min, t_max, count):
    h = (t_max - t_min) / count
    t = [(t_max - t_min) * i / count for i in range(count + 1)]

    result = [[0, 0, p_atm, p_atm, 0, p_atm, 0]]

    for i in range(count):
        result.append(Runge(t[i], result[i], h))

    return result, t, h
```

```python
tmin = 0
tmax = 2
count = 10
count2 = 20
count3 = 40
count4 = 2500
count5 = 5000
count6 = 10000
```

```python
result, t, h = hydraulic(tmin, tmax, count)
result2, t2, h2 = hydraulic(tmin, tmax, count2)
result3, t3, h3 = hydraulic(tmin, tmax, count3)
result4, t4, h4 = hydraulic(tmin, tmax, count4)
result5, t5, h5 = hydraulic(tmin, tmax, count5)
result6, t6, h6 = hydraulic(tmin, tmax, count6)
```

```python
def Runge(t, y, h):  # t - значение узла, y - значения в точке t
    # коэффициенты Рунге
    k = [[0] for _ in range(4)]

    y_temp = [0] * (size - 3)
    # значения 7 функций на следующем шаге
    y_result = [0] * size
    y_local = y.copy()

    k[0] = fun2(t, y[5], y[4])
    k[1] = fun2(t + h / 2, y[5] + h * k[0] / 2, y[4])
    k[2] = fun2(t + h / 2, y[5] + h * k[1] / 2, y[4])
    k[3] = fun2(t + h, y[5] + h * k[2], y[4])
    kn = (k[0] + 2 * k[1] + 2 * k[2] + k[3]) / 6
    # новое значение pA
    y_result[5] = y[5] + h * kn
    y[5] += h * kn

    k[0] = fun3(t, y[4], y[5], y[2], S_a)
    k[1] = fun3(t + h / 2, y[4] + h * k[0] / 2, y[5], y[2], S_a)
    k[2] = fun3(t + h / 2, y[4] + h * k[1] / 2, y[5], y[2], S_a)
    k[3] = fun3(t + h, y[4] + h * k[2], y[5], y[2], S_a)
    kn = (k[0] + 2 * k[1] + 2 * k[2] + k[3]) / 6
    # новое значение qA
    y_result[4] = y[4] + h * kn
    y[4] += h * kn

    k[0] = fun(t, y, y[4], y[6])

    for i in range(size - 3):
        y_temp[i] = y[i] + h / 2 * k[0][i]
    k[1] = fun(t + h / 2, y_temp, y[4], y[6])

    for i in range(size - 3):
        y_temp[i] = y[i] + h / 2 * k[1][i]
    k[2] = fun(t + h / 2, y_temp, y[4], y[6])

    for i in range(size - 3):
        y_temp[i] = y[i] + h * k[2][i]
    k[3] = fun(t + h, y_temp, y[4], y[6])

    for i in range(size - 3):
        y_result[i] = y[i] + h * (k[0][i] + 2 * k[1][i] + 2 * k[2][i] + k[3][i]) / 6
        y[i] = y[i] + h * (k[0][i] + 2 * k[1][i] + 2 * k[2][i] + k[3][i]) / 6

    k[0] = fun3(t, y[6], y[3], p_atm, S_b)
    k[1] = fun3(t + h / 2, y[6] + h * k[0] / 2, y[3], p_atm, S_b)
    k[2] = fun3(t + h / 2, y[6] + h * k[1] / 2, y[3], p_atm, S_b)
    k[3] = fun3(t + h, y[6] + h * k[2], y[3], p_atm, S_b)
    kn = (k[0] + 2 * k[1] + 2 * k[2] + k[3]) / 6
    # новое значение qB
    y_result[6] = y[6] + h * kn

    return y_result
```
