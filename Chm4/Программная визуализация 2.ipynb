```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# phi, ksi
phi = np.linspace(0, 2 * np.pi, 300)
ksi = np.linspace(0, np.pi, 300)

# 1 - 
x01 = 1
y01 = 2
z01 = 3
r1 = 4

x1 = x01 + r1 * np.outer(np.cos(phi), np.sin(ksi))
y1 = y01 + r1 * np.outer(np.sin(phi), np.sin(ksi))
z1 = z01 + r1 * np.outer(np.ones(np.size(phi)), np.cos(ksi))

# 2 -  
x02 = 9
y02 = 2
z02 = 3
r2 = 4

x2 = x02 + r2 * np.outer(np.cos(phi), np.sin(ksi))
y2 = y02 + r2 * np.outer(np.sin(phi), np.sin(ksi))
z2 = z02 + r2 * np.outer(np.ones(np.size(phi)), np.cos(ksi))

# 
x = np.arange(-10, 10, 0.1)
y = np.arange(-10, 10, 0.1)
z = z01

xx, yy = np.meshgrid(x, y)
z = np.sqrt((xx - x01) ** 2 + (yy - y02) ** 2 - r1 ** 2 + (xx - x02) ** 2 + (yy - y02) ** 2 - r2 ** 2)

fig = plt.figure(figsize=(10, 10))
axes = Axes3D(fig)

axes.plot_surface(x1, y1, z1, rstride=8, cstride=8, alpha=0.8, color='g')
axes.plot_surface(x2, y2, z2, rstride=8, cstride=8, alpha=0.8, color='b')

axes.contourf(xx, yy, z, zdir='z', offset=-10, levels=np.arange(0, 30, 1), cmap='Spectral', alpha=0.8)

x = []
y = []
z = []
with open("C:\\Users\\Evgeny\\ML\\Chm4+\\points.txt") as file:
        for line in file:
            a, b, c, d, e = (float(i) for i in line.split())
            x.append(a)
            y.append(b)
            z.append(c)

axes.plot(x, y, z, 'ko-')
axes.text(x[0] + 0.3, y[0] + 1, z[0] + 0.3, 'x0', color='k')
angle = 110
axes.view_init(30, angle)

axes.set_xlabel('x')
axes.set_xlim(-10, 10)
axes.set_xlabel('y')
axes.set_ylim(-10, 10)
axes.set_xlabel('z')
axes.set_zlim(-10, 10)
plt.show()
```
