```python
# окружность - построение по точкам
# x, y - центр; r - радиус
def Circ(x, y, r):
    theta = np.linspace(0, 2 * np.pi, 100)
    x1 = x + r * np.cos(theta)
    x2 = y + r * np.sin(theta)
    return x1, x2
```

```python
# number - номер системы (по программе)
# i - номер теста
def read_file(number, i):
    x = []
    y = []
    with open("C:\\Users\\Evgeny\\source\\repos\\lab4_2\\out" + str(number) + "." + str(i) + ".txt") as file:
        for line in file:
            a, b, c, d = (float(i) for i in line.split())
            x.append(a)
            y.append(b)
    return x, y
```

```python
# размер графика
plt.subplots(figsize=(13, 10))

# окружности (по точкам)
x, y = Circ(1, 2, 1)
plt.plot(x, y, 'black')
plt.text(0.2, 3.1, "(x - 1)^2 + (y - 2)^2 = 1", fontsize=12, color = 'black')
x, y = Circ(3, 2, 1)
plt.plot(x, y, 'orange')
plt.text(2.1, 3.1, "(x - 3)^2 + (y - 2)^2 = 1", fontsize=12, color = 'orange')

# первый вариант - (0, 0)
x, y = read_file(3, 1)
plt.plot(x, y, "ro-")
plt.text(x[0] + .03, y[0] + .03, "x0", fontsize=12)

# второй вариант - (0, 4)
x, y = read_file(3, 2)
plt.plot(x, y, "yo-")
plt.text(x[0] + .03, y[0] + .03, "x0", fontsize=12)

# третий вариант - (4, 3)
x, y = read_file(3, 3)
plt.plot(x, y, "co-")
plt.text(x[0] + .03, y[0] + .03, "x0", fontsize=12)

# четвёртый вариант - (4, -1)
x, y = read_file(3, 4)
plt.plot(x, y, "mo-")
plt.text(x[0] + .03, y[0] + .03, "x0", fontsize=12)

# пятый вариант - (1, 1.5)
x, y = read_file(3, 5)
plt.plot(x, y, "wo-")
plt.text(x[0] + .03, y[0] + .03, "x0", fontsize=12)

# точка пересечения
plt.text(2.05, 2.03, "A", fontsize=12, color = 'lime')

plt.grid(True)

# норма вектора F на сетке
x = np.arange(-1, 5, 0.1)
y = np.arange(-1, 5, 0.1)
xx, yy = np.meshgrid(x, y, sparse=True)
z = np.sqrt(((xx - 1) ** 2 + (yy - 2) ** 2 - 1) ** 2 +
             ((xx - 3) ** 2 + (yy - 2) ** 2 - 1) ** 2)
cs = plt.contourf(x, y, z, 300)
#plt.clabel(cs, colors='k', inline=False)
plt.colorbar()

# границы по x и y
plt.xticks(np.arange(-1, 5))
plt.yticks(np.arange(-1, 5))
plt.savefig('3')
plt.show()
```

# Две прямые пересекаются - тривиальный случай
### 1 итерация

```python
# a1, a2 - границы построения по оси OX\
# k - коэффициент в прямой, b - константа
# y = kx + b
def line(a1, a2, k, b):
    x = np.arange(a1, a2, 0.1)
    return x, k * x + b
```

```python
plt.subplots(figsize=(13, 10))

# первый вариант
x, y = read_file(1, 1)
plt.plot(x, y, "ro-")
plt.text(x[0] + .03, y[0] + .03, "x0", fontsize=12)

# первая прямая
x, y = line(-2, 0.5, 0, 2)
plt.plot(x, y, 'lime')
plt.text(-0.8, 2.1, "y = 2", fontsize=14, color = 'lime')

# вторая прямая
x, y = line(-2, 0.5, 1, 2)
plt.plot(x, y, 'k')
plt.text(-0.8, 1.1, "y = x + 2", fontsize=14, color = 'k')

# точка пересечения
plt.text(-0.03, 2.07, "A", fontsize=12, color = 'lime')

plt.grid(True)

# норма вектора F на сетке
x = np.arange(-2.5, 1.5, 0.1)
y = np.arange(0, 4, 0.1)
xx, yy = np.meshgrid(x, y, sparse=True)
z = np.sqrt((yy - 2) ** 2 + (-xx + yy - 2) ** 2)
cs = plt.contourf(x, y, z, 20)
#plt.clabel(cs, colors='k', inline=False)
plt.colorbar()

# границы по x и y
plt.xticks(np.arange(-2.5, 1.5))
plt.yticks(np.arange(0, 4))
plt.savefig('1')
plt.show()
```
